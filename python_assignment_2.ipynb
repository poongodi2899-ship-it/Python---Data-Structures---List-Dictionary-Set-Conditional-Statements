```python
# ===========================================
# PYTHON ASSIGNMENT 2
# Data Structures - List, Dictionary, Set
# & Conditional Statements
# ===========================================

print("========== LIST OPERATIONS ==========\n")

# 1. List Creation
age_list = [24, 25, 27, 28, 29]
name_list = ["Arun", "Bala", "Divya", "Kavin", "Meena"]

print("Original age_list:", age_list)
print("Original name_list:", name_list)

# 2a. Append "Yazhini"
name_list.append("Yazhini")
print("\nAfter Append:", name_list)

# 2b. Insert 26 at index 2
age_list.insert(2, 26)
print("After Insert:", age_list)

# 2c. Remove "Yazhini"
name_list.remove("Yazhini")
print("After Remove:", name_list)

# 2d. Pop last element
age_list.pop()
print("After Pop:", age_list)

# 2e. Extend list
age_list.extend([31, 30, 32])
print("After Extend:", age_list)

# 2f. Sort descending
age_list.sort(reverse=True)
print("Descending Order:", age_list)

# 2g. Max, Min, Sum
print("Maximum Age:", max(age_list))
print("Minimum Age:", min(age_list))
print("Sum of Ages:", sum(age_list))

print("\n========== LIST ACCESS ==========\n")

# 3a. First element
print("First Name:", name_list[0])

# 3b. Last element
print("Last Name:", name_list[-1])

# 3c. Elements index 2 to 4
print("Index 2 to 4:", name_list[2:5])

# 3d. Reverse order
print("Reverse Order:", name_list[::-1])

print("\n========== DICTIONARY ==========\n")

# Dictionary Creation
student_marks = {
    "Arun": 75,
    "Bala": 68,
    "Divya": 91,
    "Kavin": 84,
    "Meena": 77
}

print("Student Marks:", student_marks)

# Access mark
print("Mark of Divya:", student_marks["Divya"])

# Add Janani
student_marks["Janani"] = 80
print("\nAfter Adding Janani:", student_marks)

# Update mark
student_marks["Arun"] = 82
print("After Updating Arun:", student_marks)

# Keys, Values, Items
print("\nKeys:")
print(student_marks.keys())

print("\nValues:")
print(student_marks.values())

print("\nItems:")
print(student_marks.items())

print("\n========== SET OPERATIONS ==========\n")

# Create Set
my_set = {'a','e','i','o','u','a','a','i'}

print("my_set =", my_set)

print("\nExplanation:")
print("Set stores only unique values.")
print("Duplicate values are removed automatically.")

# Access set element
print("\nAttempting my_set[4]")

try:
    print(my_set[4])
except TypeError:
    print("Error: Sets are unordered and do not support indexing.")

# Create two sets
set1 = {1,3,5,7,9}
set2 = {2,3,5,8,10}

print("\nSet1 =", set1)
print("Set2 =", set2)

# Union
print("Union:", set1.union(set2))

# Intersection
print("Intersection:", set1.intersection(set2))

print("\n========== CONDITIONAL STATEMENTS ==========\n")

score = int(input("Enter your score (0 to 10): "))

if score < 0 or score > 10:
    print("Invalid Input! Enter a score between 0 and 10.")
elif score > 7:
    print("Above Average: Excellent work! Keep it up.")
elif score >= 4:
    print("Average: Good effort! Keep practicing, there's room for improvement.")
else:
    print("Below Average: Need to improve your performance; consistent practice will lead to better results.")
```
